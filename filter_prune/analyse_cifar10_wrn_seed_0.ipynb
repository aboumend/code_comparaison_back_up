```python
def get_data_from_csv_file(file_name):
    with open(file_name) as f:
        rows=csv.reader(f, delimiter=',')
        steps=[]
        values=[]
        i=0
        for row in rows:
            if (i==0): 
                i+=1
                continue
            else:
                steps.append(int(row[1]))
                values.append(float(row[2]))
    return steps, values     
```

```python
def get_power_stats(file, train_time):
    f=open(file, 'r')
    lines = f.readlines()
    power_list=[]
    power_avg=0


    for line in lines:
        power_list.append(int(line))
        power_avg+=int(line)

    power_avg/=len(lines)
    power_avg/=1000 #conversion en watt
    
    energy= power_avg * train_time* 60
    tr_time=len(lines)
    return power_list, power_avg, tr_time, energy
```

```python
def compute_power_cumul(pow_list):
    pow_cumul_list=[]
    pow_cum=0
    for p in pow_list:
        pow_cum+=p
        pow_cumul_list.append(pow_cum)
    return pow_cumul_list
```
